# Exploring missing data

```python
dataset["rating"].isnull().sum()
```

```python
for column in dataset.columns:
    null_per=np.mean(dataset[column].isnull())
    total = dataset[column].isnull().sum()
    if total !=0:
        print(f"percentage of null values in column {column} is {null_per}% and total number of null values is {total}")
    

```

```python
# we will replace null is equal to mean of that column(only for those columns where null values is less)
dataset["runtime"].fillna(dataset["runtime"].mean(),inplace=True)
dataset["votes"].fillna(dataset["votes"].mean(),inplace=True)
dataset["score"].fillna(dataset["score"].mean(),inplace=True)
```

```python
dataset["votes"].isnull().sum()
```

```python
dataset.dtypes
```

# change datatype of columns

```python

dataset["runtime"]=dataset["runtime"].astype("int64")
```

```python
#run only one time
dataset["budget"]=dataset.budget.round().astype('Int64')
```

```python
#run only one time
dataset["gross"]=dataset.gross.round().astype('Int64')
```

```python
# create correct year column
```

```python
# extract date and country from column released
dataset[["correct_date","correct_country"]]=dataset["released"].astype("str").str.split("(",expand=True)
```

```python
dataset[["correct_country"]]=dataset["correct_country"].astype("str").str[:-1]
```

```python
dataset[["correct_year"]]=dataset["correct_date"].astype("str").str[-5:]
```

```python
dataset["correct_year"]=dataset["correct_year"].astype("float64")
```

```python
# run only one time
dataset["correct_year"]=dataset.correct_year.round().astype('Int64')
```

```python
dataset.dtypes
```

```python
# sort values by gross
```

```python
dataset.sort_values(by=["gross"],inplace=False,ascending=False)
```

```python
dataset["company"].drop_duplicates().sort_values(ascending=True)
```

```python
# removing dulicates if any
dataset=dataset.drop_duplicates()
```

```python
dataset["gross"].isnull().sum()
```

```python
# we have to remove to duplicates to plot the scatter plot
```

```python
newdataset = dataset.dropna(subset=['budget', 'gross'])
```

```python
# to find the correlation with gross
# plot the scatter plot score  vs gross
plt.scatter(x=newdataset["budget"],y=newdataset["gross"])
plt.xlabel("Budget of movie")
plt.ylabel("Gross of movie")
plt.title("Correlation between Budget vs gross")
plt.show()
```

```python
newdataset["budget"] = newdataset["budget"].astype('float64')
newdataset["gross"] = newdataset["gross"].astype('float64')
```

```python
# plot the regplot budget vs gross using seaborn
sns.regplot(x="budget",y="gross",data=newdataset,scatter_kws={"color":"red"},line_kws={"color":"blue"})
plt.xlabel("Budget of movie")
plt.ylabel("Gross of movie")
plt.title("Correlation between Budget vs gross")
plt.show()
```

```python
# looking at correlation
```

```python
dataset.corr(method="pearson").sort_values(by="gross",ascending=False) # methods are pearson ,kendall,spearman
```

```python
# 2nd highest of gross(0.74) is budget 
# so gross is highly correlated with budget and votes
```

```python
# heatmap
correlation_matrix = dataset.corr(method="pearson").sort_values(by="gross",ascending=False)

sns.heatmap(correlation_matrix,annot=True)

plt.xlabel("Movie feature")
plt.ylabel("Movie feature")
plt.title("Correlation matrix heatmap")
plt.show()
plt.show()


```
